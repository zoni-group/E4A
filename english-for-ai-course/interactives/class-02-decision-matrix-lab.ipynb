## Step 2. Edit the safe example

Use a safe decision. The default decision is a safe example.

````python
# @title Step 2: Edit the safe example
decision_question = "Should I study in the morning or in the evening?"
option_a = "Study in the morning"
option_b = "Study in the evening"
option_c = "Study at lunchtime"
````

## Step 3. Edit the criteria and scores

Score each option from 1 (low) to 5 (high). Use whole numbers only. You
can change the criteria names. Keep them short.

````python
# @title Step 3: Edit the criteria and scores
criterion_1 = "Time available"
criterion_2 = "Energy level"
criterion_3 = "Quiet space"
criterion_4 = "Convenience"
criterion_5 = "Cost or risk"

# Score each option from 1 to 5 on each criterion.
scores_a = [5, 4, 4, 4, 5]
scores_b = [3, 3, 5, 3, 5]
scores_c = [2, 4, 2, 5, 5]

# Weights tell the matrix which criteria matter more (1 to 5).
weights = [3, 3, 2, 2, 1]
````

## Step 4. Build my matrix

Run this cell to build your decision matrix.

````python
# @title Step 4: Build my matrix
criteria = [criterion_1, criterion_2, criterion_3, criterion_4, criterion_5]
options = [option_a, option_b, option_c]
all_scores = [scores_a, scores_b, scores_c]

def clamp(value, low=1, high=5):
    return max(low, min(high, value))

clean_a = [clamp(int(v)) for v in scores_a]
clean_b = [clamp(int(v)) for v in scores_b]
clean_c = [clamp(int(v)) for v in scores_c]
clean_w = [clamp(int(v)) for v in weights]

total_a = sum(s * w for s, w in zip(clean_a, clean_w))
total_b = sum(s * w for s, w in zip(clean_b, clean_w))
total_c = sum(s * w for s, w in zip(clean_c, clean_w))

def build_table():
    header = "| Criterion | Weight | " + option_a + " | " + option_b + " | " + option_c + " |\n"
    sep = "|---|---:|---:|---:|---:|\n"
    rows = ""
    for i, c in enumerate(criteria):
        rows += "| " + c + " | " + str(clean_w[i]) + " | " + str(clean_a[i]) + " | " + str(clean_b[i]) + " | " + str(clean_c[i]) + " |\n"
    totals = "| **Weighted total** | | **" + str(total_a) + "** | **" + str(total_b) + "** | **" + str(total_c) + "** |\n"
    return header + sep + rows + totals

matrix_md = build_table()
show("**Decision question:** " + decision_question + "\n\n" + matrix_md)
````
