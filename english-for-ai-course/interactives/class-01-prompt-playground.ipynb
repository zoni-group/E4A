## Step 2. Edit the safe example

Edit the values inside the quotes. Use a safe everyday task only.

Safe task ideas:

- `"Make a simple study plan for an English test."`
- `"Write a polite email to ask my teacher a question."`
- `"Plan a small grocery list for one week."`
- `"Practice interview answers about my study goals."`
- `"Understand a short public announcement."`

````python
# @title Step 2: Edit the safe example
task = "Make a simple study plan for an English test."
context = "I can study 30 minutes a day. I want help with vocabulary and speaking."
limit_1 = "Use simple English."
limit_2 = "Do not invent facts."
limit_3 = "Do not use private information."
limit_4 = "Ask questions if information is missing."
output_format = "A 4-day study plan as a short table."
````

## Step 3. Build my ZONI CLEAR prompt

Run this cell to put your ZONI CLEAR prompt together.

````python
# @title Step 3: Build my ZONI CLEAR prompt
limits_block = "- " + limit_1 + "\n- " + limit_2 + "\n- " + limit_3 + "\n- " + limit_4
prompt_text = (
    "I need help with: " + task + "\n\n"
    "Context:\n" + context + "\n\n"
    "Limits:\n" + limits_block + "\n\n"
    "Output format:\n" + output_format + "\n\n"
    "Before answering:\n"
    "List anything that is missing or unclear."
)
show("**Your ZONI CLEAR prompt:**\n\n```text\n" + prompt_text + "\n```")
````
