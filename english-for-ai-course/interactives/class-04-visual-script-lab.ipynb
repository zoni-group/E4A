## Step 3. Build my visual description

````python
# @title Step 3: Build my visual description
valid_sources = ["self-created", "teacher-provided", "ZONI-provided course image", "open-license image"]
clean_source = source_choice if source_choice in valid_sources else "ZONI-provided course image"

description = (
    "**The image shows:** " + visual_topic + ".\n\n"
    "**The purpose is:** " + purpose + "\n\n"
    "**The text says:** \"" + text_on_visual + "\"\n\n"
    "**Source:** " + clean_source + "\n\n"
    "**Visual attribution:**\n"
    "- Title: " + visual_title + "\n"
    "- Creator: " + creator_or_author + "\n"
    "- Source: " + source_or_url + "\n"
    "- License or permission: " + license_or_permission + "\n"
    "- Changes made: " + changes_made + "\n"
    "- Attribution note: " + attribution_note
)
show(description)
````

## Step 4. Edit my 60-second speaking script

A 60-second script is around **100 to 150 words**. Read it out loud once
before saving.

````python
# @title Step 4: Edit my 60-second speaking script
script_text = (
    "Hello, my name is Student A. Today I will describe this study group flyer. "
    "The image shows a simple flyer for a free English study group at the public "
    "library. The purpose is to invite English learners of all levels. "
    "The flyer says the group meets on Saturdays at three in the afternoon. "
    "Students bring a notebook. The image uses simple words and clear colors. "
    "It does not show personal photos or private information. I like this flyer "
    "because it is short, friendly, and easy to read. After this presentation, I "
    "will share the flyer with one friend who wants to practice English with me. "
    "Thank you for listening."
)
````

## Step 5. Check the script length

````python
# @title Step 5: Check the script length
word_count = len([w for w in script_text.split() if w.strip()])

if 100 <= word_count <= 150:
    length_note = "✅ Good length: " + str(word_count) + " words (about 60 seconds)."
elif word_count < 100:
    length_note = "⚠️ Short script: " + str(word_count) + " words. Add a sentence."
else:
    length_note = "⚠️ Long script: " + str(word_count) + " words. Cut one sentence."

show("**Word count:** " + length_note)
````

## Step 8. Choose my pronunciation focus

Pick ONE focus. Read the script three times with that focus.

````python
# @title Step 8: Choose my pronunciation focus
# Choose ONE: "final consonants", "word stress", "pausing", "intonation"
pronunciation_focus = "pausing"

if pronunciation_focus == "final consonants":
    tip = "Say the last sound in each word clearly. Example: \"book\", not \"boo\"."
elif pronunciation_focus == "word stress":
    tip = "Make one syllable in each word a little longer and louder."
elif pronunciation_focus == "intonation":
    tip = "Your voice can go up at a question and down at the end of a statement."
else:
    tip = "Pause for one short breath after each sentence."

show("**My pronunciation focus:** " + pronunciation_focus + "\n\n**Tip:** " + tip)
````
